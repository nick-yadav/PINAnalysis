# Disulphide Bonds in Domains

```python
# For this we coded a python program DSBondFinder.py. In this program we find the probable Disulphide Bonds in the available 
# Domain sequences. To achieve this we programmatically Query the web page: [http://disulfind.dsi.unifi.it/index.php]with 
# the available Domain sequences. This website predicts the probable Disulphide Bonds within the submitted Amino Sequences. 
# Through the program itself we retrieve the results of the submitted query and save it. The final output file which we get
# is PINIR_Domains_DiSuplhide_Bonds.csv.
```

```python
domainDSBonds=pd.read_csv('PINIR_Domains_DiSuplhide_Bonds.csv')
```

```python
domainDSBonds.head()
```

```python
# To parse the DS Bonds we need to convert dsBonds filed into List. For this we require Abstract Syntax Trees library
```

```python
import ast # Abstract Syntax Trees
```

```python
ast.literal_eval(domainDSBonds['dsBonds'][0])## Evaluating dsBonds filed of first record to List
```

```python
ast.literal_eval(domainDSBonds['dsBonds'][0])[0]
```
