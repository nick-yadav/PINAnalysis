# Linker Organization In IRD's

```python
dom_rcl_linker=pd.read_excel('Comments/dom_rcl_linker_RefinedFields.xlsx')
```

```python
dom_rcl_linker.head() 
```

```python
dom_rcl_linker.count()
```

```python
dom_rcl_linker.loc[dom_rcl_linker.linkerID.isna(),:].count()
```

```python
dom_rcl_linker.info()
```

```python
dom_rcl_linker.linkerID.nunique()
```

```python
dom_rcl_linker
```

```python
HL=0
LH=0
NOLinker=0
for index, row in dom_rcl_linker.iterrows():
    #print(index,row['DomainType'],row['Linker_startPosition'],row['Linker_endPosition'],row['RCL_startPosition'],row['RCL_endPosition'])
    domainType=row['DomainType']
    linkerSPOS=row['Linker_startPosition']
    linkerEPOS=row['Linker_endPosition']
    rclSPOS=row['RCL_startPosition']
    rclEPOS=row['RCL_endPosition']
    
    if domainType!=3:
        #print(domainType)
        #print('RCL start Position :',rclSPOS,'/ Linker end Position :',int(linkerEPOS))
        if rclSPOS>linkerEPOS:
            print("Domain H-L Type")
            HL=HL+1
            dom_rcl_linker.loc[dom_rcl_linker.domainID==row['domainID'],'DomainType']=1
        elif rclEPOS<linkerSPOS:
            print("Domain L-H Type")
            LH=LH+1
            dom_rcl_linker.loc[dom_rcl_linker.domainID==row['domainID'],'DomainType']=2
        else:
            print("Domain without Linker")
            NOLinker=NOLinker+1
        
print('H-L Type Domains: ',HL)   
print('L-H Type Domains: ',LH) 
print('Type-3 Domains: ',NOLinker) 
    
```
