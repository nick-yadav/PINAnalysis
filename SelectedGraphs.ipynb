# 1(c) Organism wise occurence of RCL's

```python
org_rcl=pd.read_csv('Fig1/c_Nidhi_Query2_Organism_RCL_frequency_sequenceWise.csv')
```

```python
org_rcl.Organisms.nunique()
```

```python
## Extracting Records only of Top 20 Organisms with highest freuqency of PI's (uniprot_organismWise_dist.csv)
```

```python
org_rcl_top10Org=org_rcl.loc[org_rcl.Organisms.isin(['Capsicum annuum (Bell pepper)','Solanum tuberosum (Potato)','Nicotiana tabacum (Common tobacco)',
                             'Solanum lycopersicum (Tomato) (Lycopersicon esculentum)','Capsicum chinense (Scotch bonnet) (Bonnet pepper)',
                             'Capsicum baccatum (Peruvian pepper)','Nicotiana attenuata (Coyote tobacco)',
                             'Nicotiana sylvestris (Wood tobacco) (South American tobacco)','Selaginella moellendorffii (Spikemoss)',
                              'Coffea canephora (Robusta coffee)','Rosa chinensis (China rose)','Solanum chacoense (Chaco potato)',
                            'Cuscuta australis','Noccaea caerulescens (Alpine penny-cress) Thlaspi caerulescens)',
                            'Solanum americanum (American black nightshade) (Solanum nodiflorum)','Solanum nigrum (Black nightshade)',
                            'Quercus suber (Cork oak)','Panicum hallii var. hallii','Dichanthelium oligosanthes']),:]
```

```python
org_rcl_top10Org.Organisms.nunique()
```

```python
org_rcl_top10Org.head()
```

```python
orgRCLPivot=pd.pivot_table(data=org_rcl_top10Org,index='Organisms',columns='Reactive Loops (RCL)',values='count',fill_value=0)
```

```python
orgRCLPivot
```

```python
plt.figure(figsize=(20,8),dpi=300)
plt.title('Organism wise Distribution of Sequences containing a particular RCL)',pad=20,fontdict={'fontsize':15,'fontweight':'bold','fontfamily':'Arial','color':'b'})
plt.xlabel('Reactive Loops (RCL)',fontdict={'fontsize':15})
plt.ylabel('Organisms',fontdict={'fontsize':15})

sns.heatmap(orgRCLPivot,cmap='viridis')
```

```python
## The Heatmap shows the number of sequences with a particular RCL within top 20 Organism (Organisms with highest 
## freuqency of PI's). As per the given scale an increase in the intensity depicts larger number of sequences 
## with the corresponding RCL within an Organism. For e.g. organism 'Capsicum annuum (Bell pepper)' has larger no.
## of sequences with RCL's 'CPRNC' and 'CPRYC' 
```
